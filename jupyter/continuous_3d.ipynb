## EXAMPLES with restarts

### Example 08 - Bumps v2

```python
func8 = lambda x,y: hc.bumps(x,y,bump_effect=0.5) #default = 0.5
```

```python
hc.draw_graph_3d(func8,z_range=(-5.01,5.01),show=True,cmap='terrain')
```

```python
all_results, best_results, iters = hc.continious_hill_climbing_restarting_3d(func8, max_iter=10000, max_restarts=12)
```

```python
print("Steps taken for best:", len(best_results)-1)
print("Num of iterations:", iters)
print("Num of results:", len(all_results))
print("Local optimum:", func8(best_results[-1][0],best_results[-1][1]))
#print("Global optimum:", func8(0,0))
```

```python
hc.draw_graph_3d_as_imshow_with_multi_paths(func8, paths=all_results, best_path=best_results, \
                                            cmap='terrain', interpolation='bilinear', show=True)#cm.coolwarm)
```
